```python
def filtra_REG(dataframe,regiao):
  dataframe = dataframe.drop(columns=['index','coduf','codmun','Total','municipio'])
  dataframe=dataframe.query(f'regiao=="{regiao}"')
  dataframe = dataframe.groupby(by=['regiao','data','ano']).sum()
  dataframe=dataframe.reset_index()
  return dataframe
```

```python
def centroide (dfgeo,coluna):
  centroid = dfgeo.copy()
  centroid.geometry=dfgeo.centroid
  if coluna == 'casosAcumu':
    centroid['size'] = centroid[coluna]/10000
  elif coluna == 'obitosAcum':
    centroid['size'] = centroid[coluna]/1000
  return centroid
```

```python
def percentual(dataframe,ano, dataframe2,ano2):
  #dataframe 1
  mapa={"casosNovos":f"casosNovos{ano}", "obitosNovos": f"obitosNovos{ano}"}
  dataframe = dataframe.drop(columns=['index','Total','ano','data','coduf','casosAcumulado','obitosAcumulado','populacaoTCU2019'],axis=1)
  dataframe = dataframe.rename(columns=mapa)

  #dataframe 2
  mapa2={"casosNovos":f"casosNovos{ano2}", "obitosNovos": f"obitosNovos{ano2}"}
  dataframe2 = dataframe2.drop(columns=['coduf','index','Total','ano','data','municipio','populacaoTCU2019','regiao','casosAcumulado','obitosAcumulado'],axis=1)
  dataframe2 = dataframe2.rename(columns=mapa2)

  dataframe_percentual = pd.merge(dataframe,dataframe2,how = 'inner',on = 'codmun')
  dataframe_percentual=dataframe_percentual.drop('codmun',axis=1)
  dataframe_percentual=dataframe_percentual.groupby('regiao').sum()
  dataframe_percentual['casostotais']= dataframe_percentual[f"casosNovos{ano}"]+dataframe_percentual[f"casosNovos{ano2}"]
  dataframe_percentual['obitostotais']= dataframe_percentual[f"obitosNovos{ano}"]+dataframe_percentual[f"obitosNovos{ano2}"]
  dataframe_percentual['casos2020_percentual']=dataframe_percentual[f"casosNovos{ano}"]/dataframe_percentual['casostotais']*100
  dataframe_percentual['casos2021_percentual']=dataframe_percentual[f"casosNovos{ano2}"]/dataframe_percentual['casostotais']*100
  dataframe_percentual['obitos2020_percentual']=dataframe_percentual[f"obitosNovos{ano}"]/dataframe_percentual['obitostotais']*100
  dataframe_percentual['obitos2021_percentual']=dataframe_percentual[f"obitosNovos{ano2}"]/dataframe_percentual['obitostotais']*100
  return dataframe_percentual
```

```python
dados_percentuais = percentual(dados_CE_2020,'2020', dados_CE_2021,'2021')
```
